```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("insurance_data.csv")
```

```python
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.hist(bins=30, figsize=(15, 10))
plt.show()
```

```python
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()
```

```python
print(df.dtypes)
```

```python
# Convert a column to numeric, forcing errors to NaN
df['Premium'] = pd.to_numeric(df['Premium'], errors='coerce')
df['Total_Claim'] = pd.to_numeric(df['Total_Claim'], errors='coerce')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values[missing_values > 0])
```

```python
# Fill missing values with 0
df['Total_Claim'].fillna(0, inplace=True)
df['Premium'].fillna(0, inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Premium'], df['Total_Claim'])
plt.title('Total Premium vs Total Claims')
plt.xlabel('Total Premium')
plt.ylabel('Total Claims')
plt.show()
```

```python
# Check unique values in the columns
print(df['Premium'].unique())
print(df['Total_Claim'].unique())
```

```python
# Replace unwanted values with NaN
df['Premium'] = df['Premium'].replace('C', np.nan)

df['Premium'] = pd.to_numeric(df['Premium'], errors='coerce')
df['Total_Claim'] = pd.to_numeric(df['Total_Claim'], errors='coerce')
```

```python
correlation_matrix = df[['Total_Claim', 'Premium']].corr()
print(correlation_matrix)
```

```python
df.boxplot(column='Total_Claim')
plt.title('Box Plot of Total Claims')
plt.show()
```
